```
from speech_transformer import get_model
from speech_transformer import beam_search_decoding
from speech_transformer import audio_to_spec
import IPython.display as ipd
import tensorflow as tf
import librosa
import pickle
import os
```

#Model

```
#Pretrained Models :

#Subword based :
#small
CALLBACKS_PATH = 'pretrained/speech_transformer_subword_small'
#medium
#CALLBACKS_PATH = 'pretrained/speech_transformer_subword_medium'

#Character based :
#small
#CALLBACKS_PATH = 'pretrained/speech_transformer_character_small'
#medium
#CALLBACKS_PATH = 'pretrained/speech_transformer_character_medium'
```

```
with open(os.path.join(CALLBACKS_PATH, 'config.pkl'), 'br') as f:
    class Config:
        pass
    C = pickle.load(f)

model = get_model(C, os.path.join(CALLBACKS_PATH, 'checkpoints.h5'))

with open(os.path.join(CALLBACKS_PATH, 'tokenizer.pkl'), 'br') as f:
    tokenizer = pickle.load(f)
    
model.summary()
```

#Demo

```
audio = librosa.load(tf.keras.utils.get_file('OSR_us_000_0031_8k.wav','https://www.voiptroubleshooter.com/open_speech/american/OSR_us_000_0031_8k.wav'),sr=16000)[0][:80000]
ipd.display(ipd.Audio(audio, rate=16000))

pred = beam_search_decoding(audio_to_spec(audio, sr=16000), model, tokenizer, C)
print("Prediction :")
print(pred)
```
